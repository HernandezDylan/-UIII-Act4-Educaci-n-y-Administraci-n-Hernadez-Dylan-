#### Educación y Administración
* Recepcionista: Idiomas hablados, sistema de reservas, extensión telefónica, empresa, horario de entrada y tareas administrativas.

```python
# Lista vacía
recepcionistas = {}

# Agregar
def agregar():

    datos = {
        "Idiomas": input("Idiomas: "),
        "Reservas": input("Sistema de reservas: "),
        "Extension": input("Extension telefonica: "),
        "Empresa": input("Empresa: "),
        "Horario": input("Horario de entrada: "),
        "Tareas": input("Tareas administrativas: ")
    }

    recepcionistas.append(datos)

    print("Recepcionista agregada")


# Listar
def listar():

    print("\n+-------------------+-------------------+-------------+----------------+--------------+-------------------+")
    print("| Idiomas          | Reservas         | Extension  | Empresa       | Horario     | Tareas           |")
    print("+-------------------+-------------------+-------------+----------------+--------------+-------------------+")

    for r in recepcionistas:

        print(f"| {r['Idiomas']:<17} | {r['Reservas']:<17} | {r['Extension']:<11} | {r['Empresa']:<14} | {r['Horario']:<12} | {r['Tareas']:<17} |")

    print("+-------------------+-------------------+-------------+----------------+--------------+-------------------+")


# Buscar
def buscar():

    empresa = input("Empresa a buscar: ")

    for r in recepcionistas:

        if r["Empresa"] == empresa:

            print("\nRecepcionista encontrada")
            print(r)

    listar()


# Borrar
def borrar():

    empresa = input("Empresa a borrar: ")

    for r in recepcionistas:

        if r["Empresa"] == empresa:

            recepcionistas.remove(r)

            print("Recepcionista eliminada")

    listar()


# Actualizar
def actualizar():

    empresa = input("Empresa a actualizar: ")

    for r in recepcionistas:

        if r["Empresa"] == empresa:

            r["Horario"] = input("Nuevo horario: ")
            r["Tareas"] = input("Nuevas tareas: ")

            print("Datos actualizados")

    listar()


# Datos iniciales
recepcionistas.append({
    "Idiomas": "Español",
    "Reservas": "Booking",
    "Extension": "101",
    "Empresa": "Hotel Plaza",
    "Horario": "8 AM",
    "Tareas": "Llamadas"
})

recepcionistas.append({
    "Idiomas": "Español-Ingles",
    "Reservas": "Opera",
    "Extension": "102",
    "Empresa": "Hotel Norte",
    "Horario": "9 AM",
    "Tareas": "Reservas"
})

recepcionistas.append({
    "Idiomas": "Español-Frances",
    "Reservas": "ERP",
    "Extension": "103",
    "Empresa": "Hotel Luna",
    "Horario": "7 AM",
    "Tareas": "Clientes"
})

recepcionistas.append({
    "Idiomas": "Español",
    "Reservas": "HotelSoft",
    "Extension": "104",
    "Empresa": "Hotel Sol",
    "Horario": "10 AM",
    "Tareas": "Documentos"
})


# Menú
while True:

    print("\nMENU")
    print("1 Agregar recepcionista")
    print("2 Listar recepcionistas")
    print("3 Buscar recepcionista")
    print("4 Borrar recepcionista")
    print("5 Actualizar recepcionista")
    print("6 Salir")

    op = input("Selecciona opcion: ")

    if op == "1":
        agregar()

    elif op == "2":
        listar()

    elif op == "3":
        buscar()

    elif op == "4":
        borrar()

    elif op == "5":
        actualizar()

    elif op == "6":
        print("Programa finalizado")
        break

    else:
        print("Opcion incorrecta")
```
